```python
adult_education_level = pd.read_csv(SOURCE + "adult_education_level.csv")
# Combine location and time
adult_education_level['time_zone'] = adult_education_level['LOCATION']+adult_education_level['TIME'].astype(str)
# check for freq of time_zones and unique subjects and measure
adult_education_level.describe(include='all')
```

```python
features = ['Value','time_zone']
# Flag Codes all NaN values
ael_BUPPSRY = adult_education_level[adult_education_level['SUBJECT']=='BUPPSRY'].drop(columns=['SUBJECT','Flag Codes'])
ael_TRY = adult_education_level[adult_education_level['SUBJECT']=='TRY'][features]
ael_UPPSRY = adult_education_level[adult_education_level['SUBJECT']=='UPPSRY'][features]
ael_TRY_MEN = adult_education_level[adult_education_level['SUBJECT']=='TRY_MEN'][features]
ael_TRY_WOMEN = adult_education_level[adult_education_level['SUBJECT']=='TRY_WOMEN'][features]
ael_UPPSRY_MEN = adult_education_level[adult_education_level['SUBJECT']=='UPPSRY_MEN'][features]
ael_UPPSRY_WOMEN = adult_education_level[adult_education_level['SUBJECT']=='UPPSRY_WOMEN'][features]
```

```python
adult_education_subjects = (ael_BUPPSRY.merge(ael_TRY, on='time_zone', how='outer', suffixes=('', '_ael_TRY'))
                            .merge(ael_TRY_MEN, on='time_zone', how='outer', suffixes=('', '_ael_TRY_MEN'))
                            .merge(ael_TRY_WOMEN, on='time_zone', how='outer', suffixes=('', '_ael_TRY_WOMEN'))
                            .merge(ael_UPPSRY_MEN, on='time_zone', how='outer', suffixes=('', '_ael_UPPSRY_MEN'))
                            .merge(ael_UPPSRY_WOMEN, on='time_zone', how='outer', suffixes=('', '_ael_UPPSRY_WOMEN'))
                            .rename(columns={'Value': 'Value_ael_BUPPSRY','INDICATOR':'INDICATOR_ael',
                                             'MEASURE':'MEASURE_ael','FREQUENCY':'FREQUENCY_ael'}))
# rearrange column order
adult_education_subjects = adult_education_subjects[['LOCATION','INDICATOR_ael','MEASURE_ael','FREQUENCY_ael','TIME','time_zone',
                                                     'Value_ael_BUPPSRY','Value_ael_TRY','Value_ael_TRY_MEN','Value_ael_TRY_WOMEN',
                                                     'Value_ael_UPPSRY_MEN','Value_ael_UPPSRY_WOMEN']]
print(adult_education_subjects.shape)
adult_education_subjects.head()
```

```python
education_spending_subjects = (es_TRY_USD_STUDENT.merge(es_TRY_PC_GDP, on='time_zone', how='outer', suffixes=('', '_es_TRY_PC_GDP'))
                         .merge(es_PRY_NTRY_USD_STUDENT, on='time_zone', how='outer', suffixes=('', '_es_PRY_NTRY_USD_STUDENT'))
                         .merge(es_PRY_NTRY_PC_GDP, on='time_zone', how='outer', suffixes=('', '_es_PRY_NTRY_PC_GDP'))
                         .merge(es_EARLYCHILDEDU_PC_GDP, on='time_zone', how='outer', suffixes=('', '_es_EARLYCHILDEDU_PC_GDP'))
                         .merge(es_PRY_PC_GDP, on='time_zone', how='outer', suffixes=('', '_es_PRY_PC_GDP'))
                         .merge(es_EARLYCHILDEDU_USD_STUDENT, on='time_zone', how='outer', suffixes=('', '_es_EARLYCHILDEDU_USD_STUDENT'))
                         .merge(es_PRY_USD_STUDENT, on='time_zone', how='outer', suffixes=('', '_es_PRY_USD_STUDENT'))
                         .merge(es_SRY_USD_STUDENT, on='time_zone', how='outer', suffixes=('', '_es_SRY_USD_STUDENT'))
                         .merge(es_SRY_PC_GDP, on='time_zone', how='outer', suffixes=('', '_es_SRY_PC_GDP'))
                         .rename(columns={'Value': 'Value_es_TRY_USD_STUDENT','INDICATOR':'INDICATOR_es',
                                          'FREQUENCY':'FREQUENCY_es','Flag Codes':'Flag_Codes_es'}))
# rearrange column order
education_spending_subjects = education_spending_subjects[['LOCATION','INDICATOR_es','FREQUENCY_es','TIME','Flag_Codes_es','time_zone',
                                                           'Value_es_TRY_USD_STUDENT','Value_es_TRY_PC_GDP','Value_es_PRY_NTRY_USD_STUDENT',
                                                           'Value_es_PRY_NTRY_PC_GDP','Value_es_EARLYCHILDEDU_PC_GDP','Value_es_PRY_PC_GDP',
                                                           'Value_es_EARLYCHILDEDU_USD_STUDENT','Value_es_PRY_USD_STUDENT',
                                                           'Value_es_SRY_USD_STUDENT','Value_es_SRY_PC_GDP']]
print(education_spending_subjects.shape)
education_spending_subjects.describe(include='all')
# only TRY_USD_STUDENT has values. so we only use that column's values
```
